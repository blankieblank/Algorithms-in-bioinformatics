# **Реализация подсчета дистанции Левенштейна**

Необходимо реализовать подсчет дистанции Левенштейна.

Для удобства можно весь алгоритм записать в виде функции, принимающий набор аргументов:

* Первая последовательность нуклеотидов
* Вторая последовательность нуклеотидов

Функция должна возвращать дистанцию между последовательностями в численном виде.

```
# Заданные последовательности:

seq1 = TGG
seq2 = TAGC

# Возвращаемая дистанция:
distance = 2
```

````python
"""
Вычисляет расстояние Левенштейна между двумя последовательностями.
Args:
    seq1 (str): Первая последовательность.
    seq2 (str): Вторая последовательность.

Returns:
    int: Расстояние между последовательностями в численном виде.
"""

def distance(seq1, seq2) -> int:
    if not seq1 or not seq2:
        print("Введите последовательности!")

    m, n = len(seq1), len(seq2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            cost = 0 if seq1[i - 1] == seq2[j - 1] else 1
            dp[i][j] = min(dp[i - 1][j] + 1, # Удаление
                           dp[i][j - 1] + 1, # Вставка
                           dp[i - 1][j - 1] + cost) # Замена (или совпадение)
    return dp[m][n]

seq1 = "TGG"
seq2 = "TAGC"

distance = distance(seq1, seq2)

print(f"Первая последовательность: {seq1}")
print(f"Вторая последовательность: {seq2}")
print(f"Дистанция Левенштейна: {distance}")
````
